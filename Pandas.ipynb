This is a Data analysis 10 Minutes Tutorial using Pandas.
Pandas is an open-source library that provides useful APIs for data analysis. 
Series and DataFrame classes are very useful in pandas. The first one is a one dimension indexed array and the second one is like a table (2 dimensions).


```python
import pandas as Pd
import numpy as Npy
import openpyxl as pyx
```

Most of people work with data in tables formats like excel files. So Pandas can easily read and modify data in those data tables. The following sections describes some of the basic Pandas' APIs.  

## Common APIs

### Read data files

#### read_csv

Other options are: read_hdf5, read_excel.



```python
df = Pd.read_csv('Documents/astronauts.csv')
df.head()
```

### Write data files

#### to_csv

Other options are: to_hdf5, to_excel.


```python
df.to_csv('otherfile.csv')
```

Then re read the new file. It should be the same as the first file. 

```python
df = Pd.read_csv('otherfile.csv')
df.head()
```

### Idioms

- if-then
- if-then-else

### Series and Data frames


```python
ps = Pd.Series([1, 2, 3, 4])
ps
```

```python
dtf = Pd.DataFrame(         {'A': df['Unnamed: 0'],
   ...:                     'B': df['Name'],
                            'C': df['Space Flight (hr)']})

dtf
```

### Sorting



```python

dtf.sort_values(by='C', ascending=True)
```

### Grouping

Group by: 

- Spliting 
- Applying
- Combining

```python
dtf = Pd.DataFrame(         {'Name': df['Name'],
                            'Space Hours': df['Space Flight (hr)'],
                             'Space Flights': df['Space Flights'],
                            'Status': df['Status']})

dtf.groupby('Status').sum()
```

### View Data

- head
- tail
- index

```python
dtf.head()
```

```python
dtf.tail(4)
```

## EDA of the Data Set

Some new tables, missing values, outliers, plotting,etc



##### Flight hours (using head  and plotting results)

```python
# some imports to set up plotting 
import matplotlib.pyplot as plt
# pip install seaborn 
import seaborn as sns

dtfh = dtf.head()
#sns.countplot(x= 'Name', hue = 'Space Flights', data = dtfh)
dtfh.plot(x='Name', y='Space Flights', kind='bar')
```

##### Grouping by status (writing to an excel file and reading it back)

```python
dtf = Pd.DataFrame(         {'Name': df['Name'],
                            'Space Hours': df['Space Flight (hr)'],
                             'Space Flights': df['Space Flights'],
                            'Status': df['Status']})

dtf.groupby('Status').sum().to_excel('FlightHoursSummary.xlsx', sheet_name='FlightHours')
dtfx = Pd.read_excel('FlightHoursSummary.xlsx')
dtfx.head()
```

##### Missing Data

```python
df.dropna(axis = 1, how='any')
```

##### Fill missing data 

Replaced the missing data  

```python
values = {'Group':-1, 'Military Rank': 'waived', 'Military Branch': 'No army'}

df.fillna(value=values)
```

##### Grouping

```python
df.groupby(['Military Rank', 'Military Branch']).sum()
```
